# PyCity Schools Analysis

Summarizes the analysis
The analysis of School data and Student data shows that 15 total schools, with 39,170 students, varies greatly in terms of school budget and therefore per student budget. The top schools are all charter schools and the bottom schools (in terms of % overall passing) are district schools. There is not much difference in either direction when looking at reading and math scores from the 9th graders, to 12th graders. The level of comprehension that the student enters high school with, is most likely what they will graduate with. Spending ranges also does not heavily correlate to higher math or reading scores, math scores actually drop the higher the spending range is per student. And reading scores only drop about 2 points between the lowest and highest range of spending per student. The worst overall passing score is seen at the highest range of spending per student, so this shows that more money per student does not mean the student is getting a higher level of care. 

Draws two correct conclusions or comparisons from the calculations
The correlation between the top schools and the bottom schools has more to do with the total student count, compared to the per student budget. This means that the amount of students at the school has a far greater impact on the amount of students passing overall, then how much the school has to spend on each student. This most likely means there are more students getting lost per classroom in the larger district schools, compared to the charter schools with less students. The math scores have the largest gap between the bottom and top performing schools. The bottom schools have averages sitting around 77 and 76, while the top schools have average math scores at 83. When you take a look at math scores by grade, there is not much difference as 9th graders compared to 12th graders. This most likely means that the level the student enters high school as, they will graduate at that same level in terms of math comprehension. The low math scores at the district schools compared to the charter schools (looking at math scores by grade) could mean that the students filtering into the district schools came from less advantaged middle schools. Compared to the students from the charter schools most likely coming from more advantaged middle schools and even elementary schools.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data_complete['school_name'].nunique()
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete['Student ID'].nunique()
student_count
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
summary = {
    'Total Schools': [school_count],
    'Total Students': [student_count],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math_percentage],
    '% Passing Reading': [passing_reading_percentage],
    '% Overall Passing': [overall_passing_rate]}

first_summary = pd.DataFrame(summary)

# Formatting
first_summary["Total Students"] = first_summary["Total Students"].map("{:,}".format)
first_summary["Total Budget"] = first_summary["Total Budget"].map("${:,.2f}".format)
first_summary["Average Reading Score"] = first_summary["Average Reading Score"].map("{:.2f}".format)
first_summary["Average Math Score"] = first_summary["Average Math Score"].map("{:.2f}".format)
first_summary["% Passing Math"] = first_summary["% Passing Math"].map("{:.2f}%".format)
first_summary["% Passing Reading"] = first_summary["% Passing Reading"].map("{:.2f}%".format)
first_summary["% Overall Passing"] = first_summary["% Overall Passing"].map("{:.2f}%".format)

# Display the DataFrame
first_summary
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby("school_name")["math_score"].mean()
tenth_grader_math_scores = tenth_graders.groupby("school_name")["math_score"].mean()
eleventh_grader_math_scores = eleventh_graders.groupby("school_name")["math_score"].mean()
twelfth_grader_math_scores = twelfth_graders.groupby("school_name")["math_score"].mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
    "9th": ninth_grade_math_scores,
    "10th": tenth_grader_math_scores,
    "11th": eleventh_grader_math_scores,
    "12th": twelfth_grader_math_scores
})

#format data
math_scores_by_grade = math_scores_by_grade.round(2)

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
school_spending_df = per_school_summary_df.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.

school_spending_df["Per Student Budget"] = (
    school_spending_df["Per Student Budget"]
    .replace('[\$,]', '', regex=True)
    .astype(float)
)

school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"],
                                                  spending_bins, labels=labels, include_lowest=True)
                                                  

school_spending_df
```

```python
school_spending_df.dtypes

school_spending_df['Per Student Budget'] = school_spending_df['Per Student Budget'].astype(object)
```

```python
# Remove percentage signs and convert relevant columns to numeric
school_spending_df['% Passing Math'] = school_spending_df['% Passing Math'].str.rstrip('%').astype(float)
school_spending_df['% Passing Reading'] = school_spending_df['% Passing Reading'].str.rstrip('%').astype(float)
school_spending_df['% Overall Passing'] = school_spending_df['% Overall Passing'].str.rstrip('%').astype(float)

# Convert other columns to numeric, coercing errors
school_spending_df['Average Math Score'] = pd.to_numeric(school_spending_df['Average Math Score'], errors='coerce')
school_spending_df['Average Reading Score'] = pd.to_numeric(school_spending_df['Average Reading Score'], errors='coerce')

# Calculate averages for the desired columns
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

```

```python
# Combine the results into a single DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

spending_summary
```
